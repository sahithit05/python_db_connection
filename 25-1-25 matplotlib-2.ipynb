```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:\\Users\\CVR\\Desktop\\6635\\train.csv")
df.head()
```

```python
sns.barplot(data=df,x='Pclass',y='Survived',errorbar=None,palette='deep')
plt.title("Survival Rate")
plt.xlabel('Passenger class')
plt.ylabel('Survival rate')
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived',palette='deep')
plt.title("Relation between Age and Fare")
plt.show()
```

```python
x=df[['Age','Fare','Pclass','Survived']]
sns.pairplot(data=x,hue='Survived',diag_kind='kde',palette='husl')
plt.show()
```

```python
numeric_tips = df.select_dtypes(include=['number'])
correlation_matrix = numeric_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
